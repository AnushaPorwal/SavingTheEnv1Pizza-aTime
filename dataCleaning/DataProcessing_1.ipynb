```python
#food_wasted_countrywise_yearwise
countrywise_data = []
year_list = []
avg_list = []
min_list = []
max_list = []
median_list = []

for country in country_list:
    temp1 = crop_loss_filtered_2[crop_loss_filtered_2["country"] == country]
    for yr in range(1979, 2023):
        temp2 = temp1[temp1["year"] == yr]
        value = temp2["loss_percentage"].sum()
        min_val = temp2["loss_percentage"].min()
        max_val = temp2["loss_percentage"].max()
        
        if not temp2.empty:
            countrywise_data.append(country)
            year_list.append(yr)
            avg_list.append(value/ temp2.shape[0])
            min_list.append(min_val)
            max_list.append(max_val)
            median_list.append(statistics.median(temp2["loss_percentage"].to_list()))
        else:
            print(f"{country} - {yr}")
            
food_wasted_countrywise_yearwise = {"country": countrywise_data, "year": year_list, "avg_percentage": avg_list, 
                                      "min_percentage": min_list, "max_percentage": max_list, "median_percentage": median_list}
food_wasted_countrywise_yearwise = pd.DataFrame(food_wasted_countrywise_yearwise)
```
